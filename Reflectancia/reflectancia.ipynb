```python
import numpy as np
import pandas as pd
import math as m
import matplotlib.pyplot as plt
import os
```

```python
df_R_Silicio = pd.read_csv('R_Silicio.txt',skiprows=18,sep='\t',header=None)
df_R_Silicio.columns = ['Lambda(nm)','Reflectancia(%)']

df_R_IKZ2 = pd.read_csv('R_IKZ2.txt',skiprows=18,sep='\t',header=None)
df_R_IKZ2.columns = ['Lambda(nm)','Reflectancia(%)']
```

```python
# Encontrar el mínimo de reflectancia en el DataFrame IKZ2
min_reflectance_lambda = df_R_IKZ2.loc[df_R_IKZ2['Reflectancia(%)'].idxmin(), 'Lambda(nm)']

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Plot de Reflectancia vs Lambda para Silicio
plt.plot(df_R_Silicio['Lambda(nm)'], df_R_Silicio['Reflectancia(%)'], label='Silicio', color='black')

# Plot de Reflectancia vs Lambda para IKZ2
plt.plot(df_R_IKZ2['Lambda(nm)'], df_R_IKZ2['Reflectancia(%)'], label='IKZ2', color='red')

# Añadir línea vertical en 470 nm (color azul)
plt.axvline(x=470, color='blue', linestyle='--', label='470 nm')
""" plt.axvline(x=700, color='red', linestyle='--', label='700 nm')
plt.axvline(x=470, color='blue', linestyle='--', label='470 nm')
plt.axvline(x=470, color='blue', linestyle='--', label='470 nm') """
# Añadir línea vertical en el mínimo de reflectancia de IKZ2 (color verde)
plt.axvline(x=min_reflectance_lambda, color='green', linestyle='--', label=f'Mínimo Reflectancia: {min_reflectance_lambda:.0f} nm')

# Configurar etiquetas y título
plt.xlabel('Longitud de onda λ (nm)', fontsize=15)
plt.ylabel('Reflectancia (%)', fontsize=15)
plt.title('Reflectancia vs Longitud de Onda', fontsize=15)
plt.legend(fontsize = 14)
# plt.grid(True)
plt.tick_params(axis='both', which='major', labelsize=15)
# Mostrar el gráfico
plt.show()
```
